```python
class american_muscles:
  def __init__ (car, name, type, price, email, phone):
    car.name = name
    car.type = type
    car.price = price
    car.email = email
    car.phone = phone
    print(car.name)
    print(car.type)
    print(car.price)
    print(car.email)
    print(car.phone)
```

```python
obj1 = american_muscles('chevy', 'coupe', 50000, 'chevymuscle@gmail', 510245)
```

```python
class car:                                                      #Define Class
  def __init__(self, name, model, fuel, cc, cylinders):
    self.name = name
    self.model = model
    self.fuel = fuel
    self.cc = cc
    self.cylinders = cylinders
  def car_details(self):
    print(self.name)
    print(self.model)
    print(self.fuel)
    print(self.cc)
    print(self.cylinders)
```

```python
obj1 = car('Ford','Chevy','Gas','2300','8')     #Create object
```

```python
obj1.car_details()                              #Run/Print
```
