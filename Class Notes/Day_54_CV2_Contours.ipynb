```python
import cv2
import random

# Load a random image
image = cv2.imread('/content/image1.png', -1)

# Draw a circle on the image
center = (random.randint(0, image.shape[1]), random.randint(0, image.shape[0]))
radius = random.randint(10, 100)
cv2.circle(image, center, radius, (0, 255, 0), 2)

# Draw a rectangle on the image
top_left = (random.randint(0, image.shape[1]), random.randint(0, image.shape[0]))
bottom_right = (top_left[0] + random.randint(10, 100), top_left[1] + random.randint(10, 100))
cv2.rectangle(image, top_left, bottom_right, (0, 0, 255), 2)

# Draw a line on the image
start_point = (random.randint(0, image.shape[1]), random.randint(0, image.shape[0]))
end_point = (start_point[0] + random.randint(10, 100), start_point[1] + random.randint(10, 100))
cv2.line(image, start_point, end_point, (255, 0, 0), 2)

# Put text on the image
text = "This is a random shape"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image, text, (10, 20), font, 1, (0, 0, 0), 2)

# Save the image with the shape
shape = "circle" if center else "rectangle" if top_left else "line" if start_point else "text"
cv2.imwrite(f"image_with_{shape}.jpg", image)
```

# **Draw Edge Contours**

```python
img = cv.imread('/content/hand2.png')

g = cv.cvtColor(img,cv.COLOR_RGB2GRAY)

edge = cv.Canny(g, 140, 210)

contours, hierarchy = cv.findContours(edge, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)

for c in contours:
    hull = cv.convexHull(c)
    cv.drawContours(img, [hull], 0, (0, 255, 0), 5)

fig, ax = plt.subplots(1, figsize=(12,8))

plt.imshow(img)
```

# **Draw Contours**

```python
import cv2

img = cv.imread('/content/hand2.png')
# convert to RGB
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# create a binary thresholded image
_, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)
# show it
plt.imshow(binary, cmap="gray")
plt.show()
```

```python
# find the contours from the thresholded image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# draw all contours
image = cv2.drawContours(img, contours, -1, (0, 255, 0), 2)

plt.imshow(image)
plt.show()
```
