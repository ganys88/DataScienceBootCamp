```python
import pandas as pd

df = pd.read_csv('/content/2019-05-31-data.csv')
```

```python
df
```

```python
view_count=df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, c = ratio, edgecolor ='black' ) # c=ratio- giving color to ratio only not for view_count and likes
                                                               # edgecolor = 'black' gives edgecolor to the color
cbar = plt.colorbar()
cbar.set_label('ratio')

plt.xscale ('log')                       # comment it and see what difference does plt.scale makes
plt.yscale ('log')                       # comment it and see what difference does plt.scale makes

plt.xlabel('view_count')
plt.ylabel('likes')
plt.title('view_count and likes ratio')
plt.tight_layout()

# plt.scatter(x,y,s=100, c='green', marker='x'or'*' default is circle dot,
#                 edgecolor='black', linewidth = 1 this is width of the circle,
#                 alpha = 0.75 this is the softness of the circle)
```
