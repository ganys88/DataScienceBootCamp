```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/spam.tsv', delimiter = '\t')
```

```python
dataset
```

```python
dataset['label'].value_counts()
```

```python
dataset['label'].value_counts()/(len(dataset))
```

```python
dataset[dataset['label'] == 'ham']
```

```python
ham = dataset[dataset['label'] == 'ham']
```

```python
spam = dataset[dataset['label'] == 'spam']
```

```python
ham.shape, spam.shape
```

```python
ham = ham.sample(spam.shape[0])
```

```python
ham.shape, spam.shape
```

```python
df = ham.append(spam, ignore_index=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.3, random_state=0, shuffle=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
classifier = Pipeline([("tfidf", TfidfVectorizer()) , \
                       ("classifier", RandomForestClassifier(n_estimators=100))])
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred, y_test
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
